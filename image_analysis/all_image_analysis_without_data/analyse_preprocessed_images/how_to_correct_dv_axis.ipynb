We aim to find $ \mathbf{p_{\perp}} $, the component of $ \mathbf{p_0} $ that is orthogonal to $ \mathbf{SF} $, defined as:
$$
\mathbf{p_{\perp}} = \mathbf{p_0} - \mathbf{p_{\parallel}}
$$
where:
$$
\mathbf{p_{\parallel}} = \frac{\mathbf{p_0} \cdot \mathbf{SF}}{\|\mathbf{SF}\|^2} \mathbf{SF}.
$$

Instead of explicitly calculating $ \mathbf{p_{\perp}} $, the method computes:
$$
\mathbf{N_{\text{desired}}} = \mathbf{SF} \times (\mathbf{SF} \times \mathbf{p_0}).
$$

Step 1: Recall the Triple Product Identity
The cross-product $ \mathbf{SF} \times (\mathbf{SF} \times \mathbf{p_0}) $ can be expanded using the *vector triple product identity*:
$$
\mathbf{a} \times (\mathbf{b} \times \mathbf{c}) = (\mathbf{a} \cdot \mathbf{c}) \mathbf{b} - (\mathbf{a} \cdot \mathbf{b}) \mathbf{c}.
$$

Applying this to $ \mathbf{N_{\text{desired}}} $:
$$
\mathbf{N_{\text{desired}}} = \mathbf{SF} \times (\mathbf{SF} \times \mathbf{p_0}),
$$
$$
\mathbf{N_{\text{desired}}} = (\mathbf{SF} \cdot \mathbf{p_0}) \mathbf{SF} - (\mathbf{SF} \cdot \mathbf{SF}) \mathbf{p_0}.
$$

Step 2: Simplify
Let us rewrite the terms:
- $ \mathbf{SF} \cdot \mathbf{p_0} $ is the projection of $ \mathbf{p_0} $ onto $ \mathbf{SF} $.
- $ \mathbf{SF} \cdot \mathbf{SF} = \|\mathbf{SF}\|^2 $, the squared magnitude of $ \mathbf{SF} $.

Substitute these into the equation:
$$
\mathbf{N_{\text{desired}}} = \left( \frac{\mathbf{SF} \cdot \mathbf{p_0}}{\|\mathbf{SF}\|^2} \right) \mathbf{SF} - \mathbf{p_0}.
$$

Factor out $ \mathbf{p_0} $:
$$
\mathbf{N_{\text{desired}}} = -\mathbf{p_0} + \frac{(\mathbf{SF} \cdot \mathbf{p_0})}{\|\mathbf{SF}\|^2} \mathbf{SF}.
$$

Rearrange:
$$
\mathbf{N_{\text{desired}}} = -\left[ \mathbf{p_0} - \frac{(\mathbf{SF} \cdot \mathbf{p_0})}{\|\mathbf{SF}\|^2} \mathbf{SF} \right].
$$

Step 3: Recognize $ \mathbf{p_{\parallel}} $
The term:
$$
\frac{(\mathbf{SF} \cdot \mathbf{p_0})}{\|\mathbf{SF}\|^2} \mathbf{SF}
$$
is precisely $ \mathbf{p_{\parallel}} $, the projection of $ \mathbf{p_0} $ onto $ \mathbf{SF} $.

Thus:
$$
\mathbf{N_{\text{desired}}} = -\left( \mathbf{p_0} - \mathbf{p_{\parallel}} \right).
$$

But:
$$
\mathbf{p_{\perp}} = \mathbf{p_0} - \mathbf{p_{\parallel}}.
$$

So:
$$
\mathbf{N_{\text{desired}}} = -\mathbf{p_{\perp}}.
$$
